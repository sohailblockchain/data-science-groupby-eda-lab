## Setup — Import Library & Load Data
No file upload needed — data is built into the code!

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Date':     ['2025-01-01','2025-01-02','2025-01-03','2025-01-04','2025-01-05',
                 '2025-01-06','2025-01-07','2025-01-08','2025-01-09','2025-01-10',
                 '2025-01-11','2025-01-12','2025-01-13','2025-01-14','2025-01-15',
                 '2025-01-16','2025-01-17','2025-01-18','2025-01-19','2025-01-20'],
    'City':     ['Islamabad','Lahore','Islamabad','Lahore','Lahore',
                 'Islamabad','Lahore','Lahore','Lahore','Islamabad',
                 'Lahore','Lahore','Lahore','Islamabad','Islamabad',
                 'Lahore','Islamabad','Lahore','Islamabad','Islamabad'],
    'Product':  ['Keyboard','Laptop','Laptop','Keyboard','Laptop',
                 'Mouse','Monitor','Keyboard','Keyboard','Mouse',
                 'Keyboard','Laptop','Laptop','Monitor','Keyboard',
                 'Monitor','Laptop','Keyboard','Monitor','Mouse'],
    'Gender':   ['F','F','M','F','F','F','M','F','M','M',
                 'F','F','F','M','F','F','F','M','M','M'],
    'Price':    [919,572,673,580,979,76,890,559,639,250,
                 406,604,214,85,993,687,808,768,475,245],
    'Quantity': [2,2,3,5,3,2,1,5,1,3,3,5,3,3,3,3,5,5,1,4]
}

df = pd.DataFrame(data)
print('✅ Data loaded! Shape:', df.shape)
df.head()
```

---
## Step 1 — Load & Explore Data

```python
print('=== Column Names ===')
print(df.columns.tolist())

print('\n=== Data Types ===')
print(df.dtypes)

print('\n=== Basic Statistics ===')
df.describe()
```

---
## Step 2 — GroupBy
Average Price grouped by City

```python
df.groupby('City')['Price'].mean()
```

---
## Step 3 — Aggregation
Multiple statistics at once: mean, max, min, **count**

```python
df.groupby('City')['Price'].agg(['mean', 'max', 'min', 'count'])
```

---
## Step 4 — Multi-Column GroupBy
Group by City AND Gender together

```python
df.groupby(['City', 'Gender'])['Price'].mean()
```

---
## Step 5 — Pivot Table
Cross-tab view: City (rows) vs Gender (columns)

```python
pd.pivot_table(df, values='Price', index='City', columns='Gender', aggfunc='mean')
```

---
## Bonus — Bar Chart Visualization

```python
df.groupby('City')['Price'].mean().plot(kind='bar', color=['steelblue', 'orange'])
plt.title('Average Price by City')
plt.ylabel('Price (Rs)')
plt.xlabel('City')
plt.tight_layout()
plt.show()
```
