### What is EDA?
EDA = **Exploratory Data Analysis** — we explore a dataset BEFORE building any model.
We answer: *What does this data look like? What problems exist? What patterns can we find?*

### The 6 Steps we will follow:
1. Load & Preview
2. Data Info & Types
3. Missing Values
4. Descriptive Statistics
5. Grouping & Category Analysis
6. Visualization (Charts)

## Cell 1 — Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print('✅ All libraries loaded successfully!')
print('Pandas version:', pd.__version__)
```

---
## Cell 2 — Load the Titanic Dataset
Seaborn has Titanic built-in — no file upload needed!

```python
# Load Titanic dataset from seaborn (built-in, no download needed)
df = sns.load_dataset('titanic')

print('✅ Dataset loaded!')
print('Shape (rows, columns):', df.shape)
print('\nFirst 5 rows:')
df.head()
```

---
## Cell 3 — Step 1: Preview the Data
Look at top and bottom rows to understand the structure

```python
# First 5 rows
print('=== FIRST 5 ROWS ===')
print(df.head())

# Last 5 rows
print('\n=== LAST 5 ROWS ===')
print(df.tail())

# Shape
print('\n=== DATASET SHAPE ===')
print(f'Rows: {df.shape[0]}')
print(f'Columns: {df.shape[1]}')
print('\nColumn names:', df.columns.tolist())
```

---
## Cell 4 — Step 2: Data Info & Column Types
Understand data types and which columns have null values

```python
# Full info: types, non-null counts, memory
print('=== DATA INFO ===')
df.info()

print('\n=== DATA TYPES ===')
print(df.dtypes)

print('\n=== UNIQUE VALUES PER COLUMN ===')
print(df.nunique())
```

---
## Cell 5 — Step 3: Missing Values Analysis
Find columns with missing data and visualize them

```python
# Count missing values
print('=== MISSING VALUES COUNT ===')
print(df.isnull().sum())

# Missing values as percentage
print('\n=== MISSING VALUES PERCENTAGE ===')
missing_pct = (df.isnull().sum() / len(df)) * 100
print(missing_pct.round(1))

# Bar chart of missing values
plt.figure(figsize=(12, 4))
df.isnull().sum().plot(kind='bar', color='coral', edgecolor='white')
plt.title('Missing Values per Column', fontsize=14)
plt.ylabel('Number of Missing Values')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print('\n--- Decision ---')
print('Age (19.9% missing)  → Fill with median age')
print('Deck (77.2% missing) → Drop this column (too many missing)')
print('Embarked (0.2%)      → Fill with most common value (mode)')
```

---
## Cell 6 — Handle Missing Values
Fix missing data before analysis

```python
# Fill age with median
df['age'].fillna(df['age'].median(), inplace=True)

# Fill embarked with mode (most common)
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)

# Verify
print('=== MISSING VALUES AFTER CLEANING ===')
print(df.isnull().sum())
print('\n✅ Age and Embarked fixed!')
```

---
## Cell 7 — Step 4: Descriptive Statistics
Mean, median, min, max for all numeric columns

```python
# Full statistics table
print('=== DESCRIPTIVE STATISTICS ===')
display(df.describe())

# Key individual stats
print('\n=== KEY STATISTICS ===')
print(f'Average Age:        {df["age"].mean():.1f} years')
print(f'Median Age:         {df["age"].median():.1f} years')
print(f'Average Fare:       ${df["fare"].mean():.2f}')
print(f'Max Fare:           ${df["fare"].max():.2f}')
print(f'Survival Rate:      {df["survived"].mean()*100:.1f}%')
print(f'Total Passengers:   {len(df)}')
print(f'Total Survived:     {df["survived"].sum()}')
```

---
## Cell 8 — Step 5: Grouping & Category Analysis
Find patterns — who survived and why?

```python
# Survival by Gender
print('=== SURVIVAL RATE BY GENDER ===')
gender_survival = df.groupby('sex')['survived'].mean() * 100
print(gender_survival.round(1).to_string())

# Survival by Passenger Class
print('\n=== SURVIVAL RATE BY PASSENGER CLASS ===')
class_survival = df.groupby('pclass')['survived'].mean() * 100
print(class_survival.round(1).to_string())

# Survival by Gender AND Class (multi-group)
print('\n=== SURVIVAL RATE BY GENDER AND CLASS ===')
print(df.groupby(['sex', 'pclass'])['survived'].mean().round(2) * 100)

# Survival by Embarkation port
print('\n=== SURVIVAL RATE BY EMBARKATION PORT ===')
print(df.groupby('embarked')['survived'].mean().round(2) * 100)

# Count passengers per class
print('\n=== PASSENGER COUNT BY CLASS ===')
print(df['pclass'].value_counts())
```

---
## Cell 9 — Step 6: Visualization (4 Key Charts)
Charts make patterns visible instantly

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Chart 1: Survival Count
df['survived'].value_counts().plot(
    kind='bar', ax=axes[0,0],
    color=['coral', 'steelblue'], edgecolor='white')
axes[0,0].set_title('Survival Count (0=No, 1=Yes)', fontweight='bold')
axes[0,0].set_xlabel('Survived')
axes[0,0].set_ylabel('Count')
axes[0,0].tick_params(rotation=0)

# Chart 2: Survival Rate by Gender
df.groupby('sex')['survived'].mean().plot(
    kind='bar', ax=axes[0,1],
    color=['steelblue', 'coral'], edgecolor='white')
axes[0,1].set_title('Survival Rate by Gender', fontweight='bold')
axes[0,1].set_ylabel('Survival Rate')
axes[0,1].tick_params(rotation=0)

# Chart 3: Age Distribution
df['age'].plot(
    kind='hist', bins=30, ax=axes[1,0],
    color='steelblue', edgecolor='white')
axes[1,0].set_title('Age Distribution of Passengers', fontweight='bold')
axes[1,0].set_xlabel('Age')
axes[1,0].set_ylabel('Count')

# Chart 4: Survival Rate by Class
df.groupby('pclass')['survived'].mean().plot(
    kind='bar', ax=axes[1,1],
    color=['gold', 'silver', 'peru'], edgecolor='white')
axes[1,1].set_title('Survival Rate by Passenger Class', fontweight='bold')
axes[1,1].set_ylabel('Survival Rate')
axes[1,1].set_xlabel('Class (1=First, 2=Second, 3=Third)')
axes[1,1].tick_params(rotation=0)

plt.suptitle('Titanic EDA — Key Charts', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

---
## Cell 10 — Bonus: Correlation Heatmap
Which numeric columns are related to each other?

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10, 7))
sns.heatmap(
    numeric_df.corr(),
    annot=True, fmt='.2f',
    cmap='coolwarm', center=0,
    linewidths=0.5, square=True,
    cbar_kws={'shrink': 0.8}
)
plt.title('Correlation Heatmap — Titanic Dataset', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print('\n=== CORRELATION WITH SURVIVED ===')
print(numeric_df.corr()['survived'].sort_values(ascending=False).round(3))
```
